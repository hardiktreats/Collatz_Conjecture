#If the initial and final states are as below, find the value of Heuristic function, by taking
#(i) Euclidean Distance
#(ii) Manhattan Distance
#(iii) Minkowski Distance

```python
import numpy as np
import math
hardik = np.array([[2,0,3], [1,8,4], [7,6,5]])
dhanjal = np.array([[1,2,3], [8,0,4], [7, 6, 5]])
hd_m_3329=0
hd_e_3329=0
hd_mk_3329=0
t=3
for i in range(3):
  for j in range(3):
    for p in range(3):
      for q in range(3):
        if hardik[i][j]==dhanjal[p][q]:
          hd_m_3329=hd_m_3329 + abs(p-i) + abs(q-j)
          hd_e_3329=hd_e_3329+math.sqrt(((p-i)**2)+((q-j)**2))
          hd_mk_3329=hd_mk_3329+float(((abs(p-i)**t)+(abs(q-j)**t))**(1/t))
print("Manhattan distance: ",hd_m_3329)
print("Euclidean distance: ",hd_e_3329)
print("Minkowski distance: ",hd_mk_3329)
```

n as compared to the goal node need to be considered as the heuristic function. You need to use
BestFirst Search algorithm.

```python

import copy
hardik_queueState=[]
hardik_visitedState=[]
ctr=0  #variable to store how many times function puzzle is called
class node:
  def __init__(hardik_self, node, parent, heuristic):
    hardik_self.node=node
    hardik_self.parent=parent
    hardik_self.heuristic=heuristic
def heuristic(hardik_currState, hardik_goalState): #function to count number of mismatches and return the count
  ctr=0
  for i in range(3):
    for j in range(3):
      if hardik_currState[i][j]!=hardik_goalState[i][j]:
        ctr=ctr+1
  return ctr
def indexOfZero(hardik_currState): #function to find index of 0
  for i in range(3):
    for j in range(3):
      if hardik_currState[i][j]==0:
        return [i, j]
def up(s): #function to move 0 upwards
  hardik_currState=copy.deepcopy(s)
  [i, j]=indexOfZero(hardik_currState)
  if i>0:
    hardik_currState[i][j], hardik_currState[i-1][j] = hardik_currState[i-1][j], hardik_currState[i][j]
  return hardik_currState
def down(s): #function to move 0 downwards
  hardik_currState=copy.deepcopy(s)
  [i, j]=indexOfZero(hardik_currState)
  if i<2:
    hardik_currState[i][j], hardik_currState[i+1][j] = hardik_currState[i+1][j], hardik_currState[i][j]
  return hardik_currState
def right(s): #function to move 0 towards right
  hardik_currState=copy.deepcopy(s)
  [i, j]=indexOfZero(hardik_currState)
  if j<2:
    hardik_currState[i][j], hardik_currState[i][j+1] = hardik_currState[i][j+1], hardik_currState[i][j]
  return hardik_currState
def left(s): #function to move 0 towards left
  hardik_currState=copy.deepcopy(s)
  [i, j]=indexOfZero(hardik_currState)
  if j>0:
    hardik_currState[i][j], hardik_currState[i][j-1] = hardik_currState[i][j-1], hardik_currState[i][j]
  return hardik_currState
def notInVisitedState(hardik_currState):
  for state in hardik_visitedState:
    if hardik_currState == state.node:
      return False
  return True
def puzzle(s, g): #iterative function to solve the puzzle
  global ctr
  if s==g:
    return
  new=s
  if notInVisitedState(up(new)) and up(new) != new:
    hardik_queueState.append(node(up(new), new, heuristic(up(new), g)))
  if notInVisitedState(down(new)) and down(new) != new:
    hardik_queueState.append(node(down(new), new, heuristic(down(new), g)))
  if notInVisitedState(left(new)) and left(new) != new:
    hardik_queueState.append(node(left(new), new, heuristic(left(new), g)))
  if notInVisitedState(right(new)) and right(new) != new:
    hardik_queueState.append(node(right(new), new, heuristic(right(new), g)))
  hardik_queueState.sort(key=lambda node: node.heuristic)
  state=hardik_queueState[0]
  hardik_visitedState.append(state)
  newState=hardik_queueState.pop(0).node
  ctr=ctr+1
  puzzle(newState, g)
path=[]
def pathFinder(hardik_visitedNode):
  if hardik_visitedNode.parent=="NULL":
    return
  path.append(hardik_visitedNode.node)
  for state in hardik_visitedState:
    if state.node==hardik_visitedNode.parent:
      pathFinder(state)
def main():
  s=[[2,0,3],[1,8,4],[7,6,5]]
  hardik_goalState=[[1,2,3],[8,0,4],[7,6,5]]
  hardik_queueState.append(node(s, "NULL", heuristic(s, hardik_goalState)))
  hardik_visitedState.append(node(s, "NULL", heuristic(s, hardik_goalState)))
  puzzle(s, hardik_goalState)
  print("We've reached the goal state.")
  pathFinder(hardik_visitedState[-1])
  path.reverse()
  print("Number of moves:", len(path), "\nPath:", [s]+path)
  print('No. of iterations=', ctr)

if __name__=="__main__":
  main()

```

#Ques 5: Solve this given problem using Uniform cost search. A is the initial state and G is the goal state

```python
import copy
graph = {
  'S':['A','B','C'],
  'A':['G'],
  'B':['G'],
  'C':['G'],
  'G':[]
}
hardik_cost = {
  ('S','A'):1,
  ('S','B'):5,
  ('S','C'):15,
  ('A','G'):10,
  ('B','G'):5,
  ('C','G'):5,
}
hardik_totCost=0
hardik_path=[]
hardik_paths=[]
def problem(node, end):
  global hardik_totCost, hardik_paths
  if node==end:
    newpath=copy.deepcopy(hardik_path)
    hardik_paths.append([hardik_totCost, newpath])
    hardik_path.clear()
    hardik_path.append('A')
    hardik_totCost=0
    return
  for neighbour in graph[node]:
    hardik_totCost+=hardik_cost[node, neighbour]
    hardik_path.append(neighbour)
    problem(neighbour, end)
  hardik_paths.sort()
hardik_path.append('A')
problem('A', 'G')
print('The shortest path is: ', end='')
for node in hardik_paths[0][1]:
  if node!=hardik_paths[0][1][-1]:
    print(node,'->',sep='',end='')
  else:
    print(node,end='')
print(' with tot_cost =', hardik_paths[0][0])
```
